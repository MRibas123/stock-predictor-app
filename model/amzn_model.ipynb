```python
from pandas_datareader import data
```

```python
amzn = data.DataReader("AMZN", 
                       start='2000-1-1', 
                       end='2020-12-31', 
                       data_source='yahoo')
```

```python
amzn = amzn.reset_index()
```

```python
amzn['Year'] = amzn['Date'].dt.year
amzn['month'] = amzn['Date'].dt.month
amzn['day'] = amzn['Date'].dt.day
```

```python
def quarter(row):
    if row['month'] == 1 or row['month'] == 2 or row['month'] == 3:
        return 1
    elif row['month'] == 4 or row['month'] == 5 or row['month'] == 6:
        return 4
    elif row['month'] == 7 or row['month'] == 8 or row['month'] == 9:
        return 7
    else:
        return 10

amzn['quarter'] = amzn.apply(quarter, axis =1)
```

```python
import matplotlib.pyplot as plt
plt.plot(amzn['Adj Close'])
plt.show()
```

```python
import pandas as pd

data = pd.read_csv('data_usa.csv', index_col=0)

data_amzn = data[data['year'] >= 2000 ]
data_amzn
```

```python
amzn_final = pd.merge(left = data_amzn.reset_index(drop=True), right = amzn, left_on = 'month', right_on = 'quarter', how= 'left')
```

```python
amzn_final = amzn_final[['Date','GDP','growth_rate','jobs','growth_jobs','High','Low','Open','Close','Volume','Adj Close','quarter','Year','month_y','day']].set_index('Date')

amzn_final
```

```python
X = amzn_final.drop(['GDP','jobs','Close','Volume'], axis = 1)
y = amzn_final['Close']

X_train = X[:105416]
y_train = y[1:105417]

X_test = X[105416:len(X)-1]
y_test = y[105417:]
```

```python
import xgboost

xgb_reg = xgboost.XGBRegressor()
xgb_reg.fit(X_train, y_train)
display(xgb_reg.score(X_test,y_test))
display(xgb_reg.score(X_train,y_train))
```

```python
pred = xgb_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score


r2_score(y_test, pred)
```

```python
from sklearn.model_selection import cross_validate

results = cross_validate(xgb_reg,X_train, y_train, cv = 10)
```

```python
results['test_score'].mean()
results['test_score']
```
