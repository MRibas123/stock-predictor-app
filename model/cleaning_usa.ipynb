# GDP data

```python
import pandas as pd

gdp = pd.read_csv('GDP.csv')
```

```python
gdp
```

```python
gdp_dict= {
    '2020-04-01': 19520.114,
    '2020-07-01': 21170.252,
    '2020-10-01':21487.896
        }

gdp2020 = pd.DataFrame.from_dict(gdp_dict, orient='index').reset_index()

gdp2020.columns = ['DATE' , 'GDP']
```

```python
real_gdp = pd.concat([gdp,gdp2020]).reset_index(drop = True)
```

```python
real_gdp.to_csv('gdp.csv')
```

```python
gdp = pd.read_csv('GDP.csv' ,parse_dates=True, index_col='DATE').drop(['Unnamed: 0'],axis =1)

gdp
```

# Labor Data

```python
labor = pd.read_csv('labor.csv',index_col='Year')
labor = labor.drop(['Feb', 'Mar','May','Jun','Sep', 'Aug', 'Nov','Dec'],axis =1).reset_index()
```

```python
labor
```

```python
labor = labor.melt(id_vars=["Year"],value_vars =['Jan','Apr','Jul','Oct'],value_name="jobs")
```

```python
labor
```

```python
labor

labor['DATE'] = pd.to_datetime(labor[['Year', 'month']].assign(DAY=1))
```

```python
def month_num(row):
    if row['variable'] == 'Jan':
        return 1
    elif row['variable'] == 'Apr':
        return 4
    elif row['variable'] == 'Jul':
        return 7
    elif row['variable'] == 'Oct':
        return 10
    
labor['month'] = labor.apply(month_num,axis=1)

labor
```

```python
labor = labor[labor['Year'] != 2021]
labor
```

```python
labor = labor.sort_values(by=['Year','month']).reset_index(drop=True).drop(['variable'], axis =1)
```

```python
labor.set_index('DATE',inplace = True)
```

```python
labor.drop(['Year','month'],axis = 1,inplace= True)
```

```python
labor = pd.DataFrame(pd.to_numeric(labor['jobs']))
labor['growth_jobs'] = labor.pct_change()*100
```

```python
labor
```

```python
final_data = gdp.join(labor, on= 'DATE')
```

```python
final_data.to_csv('data_usa.csv')
```
